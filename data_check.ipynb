### 3. Jolpica API Query

```python
import requests

url = "https://api.jolpi.ca/ergast/f1/2024/driverstandings/"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    standings = data['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
    
    print(f"HTTP Status Code: {response.status_code}")
    print(f"Number of records returned: {len(standings)}")
    print("\nFirst 3 driver entries:")
    
    for i in range(3):
        driver = standings[i]['Driver']
        points = standings[i]['points']
        print(f"- {driver['givenName']} {driver['familyName']} ({driver['nationality']}): {points} points")
else:
    print(f"Error: {response.status_code}")
```
